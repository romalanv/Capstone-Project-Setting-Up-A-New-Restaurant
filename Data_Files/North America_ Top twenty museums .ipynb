```python
#Import relevant Libraries

import requests 
import urllib

import pandas as pd

from bs4 import BeautifulSoup
```

```python
# Using BeautifulSoup download the relevant data 

url = urllib.request.urlopen("https://en.wikipedia.org/wiki/List_of_most_visited_museums")
soup = BeautifulSoup(url, 'lxml')
print(soup.prettify())
```

```python
table = soup.find('table', {'class': 'wikitable sortable'})
print(table)
```

```python
table_rows = table.find_all("tr")
Col = []
for tr in table_rows:
    td = tr.find_all("td")
    row = [tr.text for tr in td]
    Col.append(row)
Col[0] # Checking the first value in the list.
```

```python
#Delete the first entry as it is empty 
Col.pop(0)
Col[0]
```

```python
#Create a new dataframe with the columns
column_names = ['Name', 'Country Flag/City', 'Visitors per year','Year Reported']
topmuseums = pd.DataFrame(columns = column_names, data = Col)
topmuseums.head() 
```

```python
#Since the Ref column values have an extra character '\n', remove it by using the rstrip() method
topmuseums['Year Reported'] = topmuseums['Year Reported'].apply(lambda x: x.rstrip('\n'))
topmuseums['Year Reported'].head()
```

```python
topmuseums.sort_values(by=['Visitors per year'], ascending=True)
```
