```python
df = TopMuseums
```

```python
Google_address = df.Name.unique()
Google_address
```

```python
#Set up the for loop and search for each lat an Lng for each address 
# Define Geocode and Latlong above to enable the For Loop 
#Setup an exception Index Error in an event the address in not found 
#Write the results to the Dataframe storing the relevant infromation 

for address in Google_address:
    single_address = address
    try:
        geocode = gmaps.geocode(single_address)[0]
        latlong = geocode['geometry']['location']
    except IndexError:
        latlog = {'lat': None,
                  ' lng': None}
    
    df.loc[df.Name == address, 'lat'] = latlong['lat']
    df.loc[df.Name == address, 'long'] = latlong['lng']
    
```
